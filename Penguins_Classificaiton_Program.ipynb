## Data Preparation

```python
from sklearn import preprocessing
def prep_data(df):
    
    '''
    Prepares data by eliminating unnecessary columns and 
    converting string data to numerical data.
    Input: A dataframe to be cleaned
    Returns: A cleaned dataframe.
    '''
    df = df.copy()
    #remove the columns that are not characteristics of penguins
        #studyName comments, individual id, and sample number are for study, 
        #not characteristics of penguins
        #Stage is the age of the penguin, not unique to a species, 
        #Date egg seems individual too
        
    df = df.drop(labels = ['studyName','Sample Number','Stage',
                           'Individual ID','Date Egg','Comments'], axis = 1)
    df = df[(df['Sex'] != '.')]
    df = df.dropna()
    
    #Using the preprocessing module from sklearn convert all string data to numerical data
    le=preprocessing.LabelEncoder()
    df['Species']=le.fit_transform(df['Species'])
    df['Island']=le.fit_transform(df['Island'])
    df['Sex']=le.fit_transform(df['Sex'])
    df['Region']=le.fit_transform(df['Region'])
    df['Clutch Completion']=le.fit_transform(df['Clutch Completion'])
    
    return(df)
```

```python
df = prep_data(penguins) # prep the penguins
```

```python
#Selected predictor variables and set target variables
def split_data(df):
    '''
    Splits a penguins data frame into predictor and target variables.
    Input: A data frame to be split.
    Returns: 2 dataframes, one with predictor variables and one with the target variable
    '''
    X = df[['Island','Culmen Length (mm)','Culmen Depth (mm)']]
    y = df['Species']
    return X,y
```

```python
from sklearn.model_selection import train_test_split
#split the data into a training and testing test
train, test = train_test_split(df, test_size = .2)
X_train, y_train = split_data(train)
X_test, y_test = split_data(test)
```

#### Plotting Function 
This will be used to plot regions for the following machine learning models.

```python
def plot_regions(c, X, y,island):
    '''
    Plots the decision regions for a model of penguins
    Inputs: c the model to plot
            X, the predictor variables
            y, the target variable
            island, the island of penguins being plotted
    Return: 3 plots, one for each island that shows the decision regions
    for a penguins. Each penguin is it's own color matching it's region
    Plots are labeled by island and have a legend for the penguins.
    
    '''  
    
    #create a dictionary of colors, labels for the penguins
    penguin_data={0: ['blue', 'Adelie'],1:['green','Chinstrap'],2:['red','Gentoo']} 
    #create a dictionary of titles for the islands
    island_names = {0: 'Biscoe',1:'Dream',2:'Torgersen'}
    
    # for convenience, give names to the two 
    # columns of the data
    mask = X['Island']==island
    x0 = X['Culmen Length (mm)'][mask]
    x1 = X['Culmen Depth (mm)'][mask]
    
    # create a grid
    grid_x = np.linspace(x0.min(),x0.max(),501)
    grid_y = np.linspace(x1.min(),x1.max(),501)
    xx, yy = np.meshgrid(grid_x, grid_y)
    
    # extract model predictions, using the 
    # np.c_ attribute to join together the two parts of the grid. 
    # array.ravel() converts an multidimensional
        # array into a 1d array
    #We make an island array so that prediction parameters match
    #we use array.reshape() to turn the resulting predictions p 
        # back into 2d
    
    XX = xx.ravel()
    YY = yy.ravel()
    Island = np.ones(XX.shape)*island
    XY = np.c_[Island, XX, YY]
    
    p = c.predict(XY)
    p = p.reshape(xx.shape)
    
    # create the plot
    fig, ax = plt.subplots(1)
    
    # use contour plot to visualize the predictions
    ax.contourf(xx, yy, p, alpha = 0.2, cmap = 'jet')
    
    # plot the data, each penguin species seperately
    for species in [0,1,2]:
        ax.scatter(x0[y==species], x1[y==species],  c = penguin_data[species][0],
                 label = penguin_data[species][1], alpha = .5)
    
    #add axis labels
    ax.set(xlabel = "Culmen Length (mm)", 
           ylabel = "Culmen Depth (mm)",
          title = island_names[island])
    #add legend
    ax.legend()
```

## Decision Tree Model

```python
from sklearn import tree
# tried with max_depth=2
T = tree.DecisionTreeClassifier(max_depth =  2)
T.fit(X_train, y_train)
# find the score of the train and test sets
T.score(X_train,y_train), T.score(X_test,y_test)
```

We created a initial model of the Decision Tree. We started with max_depth 2 as a random example of a tree in order to preview the score. 

```python
fig, ax = plt.subplots(1, figsize = (20, 20))
# plot the decision tree
p = tree.plot_tree(T, filled = True, feature_names = X_train.columns)
```

A decision tree is a model that is imported from the sklearn module. It behaves similar to a flow chart. The model follows through each step based on the characteristics examined in each box. It follows a path through the tree. The conclusion of the model occurs in the small boxes at the end of each branch. The value line in the boxes describes the number of penguins from each species that are sorted into that box. The first line in every box states the decision that is to be made. If the penguin conforms to the characteristic it follows one path and if not it follows the other path. 

```python
# using cross validation to find the best depth
from sklearn.model_selection import cross_val_score
# set as negative infinity so it is always overwritten
best_score=-np.inf
N=30 #largest max depth
scores=np.zeros(N)

# loop over the depths and find the best depth
for d in range(1,N+1):
    # try different depth values on the tree
    T=tree.DecisionTreeClassifier(max_depth=d)
    scores[d-1]=cross_val_score(T,X_train,y_train,cv=5).mean()
    if scores[d-1]>best_score:
        # update the value of d
        best_depth=d
        best_score=scores[d-1]
# print the best_depth and best_score        
best_depth,best_score
```

The cell above does cross validation. We used this to find the best depth for the tree. Cross validation splits the dataset up into different sections and tests the model on different training sets. The for loop tests cross validation on different depths. This ensures that we don't choose a depth with over-fitting. Over-fitting occurs when the model is too specific to the training set and is not as accurate on the test set. Because cross-validation scores on a "test" set it can prevent over fitting. We found that the best depth is 4 and will be using that going forward.

```python
# overwrite max_depth as best_depth
T = tree.DecisionTreeClassifier(max_depth =  best_depth)
T.fit(X_train, y_train)
# now score it with the best_depth
T.score(X_train,y_train), T.score(X_test,y_test)
```

```python
# print the model with the best_depth value instead
fig, ax = plt.subplots(1, figsize = (20, 20))
p = tree.plot_tree(T, filled = True, feature_names = X_train.columns)
```

Our model used a depth of 4. This was found from cross validation(see above). There are more decisions made in this tree than the one with max_depth of 2, as can be seen by the longer paths. It has a higher accuracy than the depth 2 tree. The decision tree helps to categorize the species of penguins based of of specific conditions. Because more conditions are evaluated in the higher depth tree, it is more accurate.

```python
#Set up cross evaluation to see how it works now
cv_score = cross_val_score(LR, X_train, y_train, cv=5).mean()
cv_score
```

```python
#set the initial best score as -infinity
    #Every score should be better so will not affect results
best_score = -np.inf
#loop through potential values of C and test using cross validation
for C in range(1, 15):
    LR = LogisticRegression(max_iter = 1000, C = C)
    LR.fit(X_train, y_train)
    score = cross_val_score(LR, X_train, y_train, cv = 5).mean()
    if best_score < score:
        best_score = score
        best_C = C
        
print(best_score, best_C) #print the found best score and best C
```

The cross validation on the Logistic Regression model determines the complexity of the model, rather than the max_depth as seen in the tree case. The system itself works the same as the tree case. 

```python
#Create a Logistic Regression Model on the best C value
LR=LogisticRegression(max_iter = 1000, C = best_C)
LR.fit(X_train, y_train)
LR.score(X_train, y_train)
```

```python
from sklearn import svm
#make the SVM model
SVM = svm.SVC(gamma = 1)
SVM.fit(X_train,y_train) # fit it to data
SVM.score(X_train, y_train) #test the model 
```

We created a beginning SVM model with a gamma value of 1. SVM models inputs data points and then draws a line that separates the data points. These lines separate the data into categories. Depending on which side of the line the penguin data point falls in, the penguin is characterized as one of the species. 

```python
N = 30 #largest complexity gamma value
#intializing lists to store scores
scores = np.zeros(N) 
best_score = -np.inf

#Loop over different gamma values
for g in range(1, N+1):
    gamma_choice = g/10
    SVM = svm.SVC(gamma = gamma_choice)
    #Compute cross val score of model with gamma value
    scores[g-1] = cross_val_score(SVM, X_train, y_train, cv =5).mean()
    #Update best_score and best_gamma if score is better than the current best
    if scores[g-1]>best_score:
        best_score = scores[g-1]
        best_gamma = gamma_choice
        
best_gamma, best_score #print best score and best gamma
```

Cross validation in the SVM case determines the best gamma value. The gamma value determines the complexity. The cross validation system works similar to the tree and Linear Regression cases. 

```python
#Making the SVM model with the optimal gamma
SVM = svm.SVC(gamma =best_gamma) 
SVM.fit(X_train, y_train)
print('Training score: ', SVM.score(X_train, y_train))
#scoring the model on the test set
print('Test score: ' , SVM.score(X_test, y_test))
```

According to the best gamma of .8, the score is about 98.4% accurate when we apply it to the test data. There is some decrease from the training score to the test score. The difference is only about 1% which is expected. The lack of a large gap indicates that overfitting is unlikely.
